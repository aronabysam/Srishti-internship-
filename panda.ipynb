```python
import pandas as pd

df1 = pd.DataFrame({
    'key': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'key': [1, 2, 3],
    'Age': [25, 30, 35]
})

pd.merge(df1, df2, on='key')

```

```python
df1.join(df2, lsuffix='_left', rsuffix='_right')
```
